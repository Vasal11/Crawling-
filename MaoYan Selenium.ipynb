```python
import openpyxl as xl
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
def driver_start(url):
    driver = webdriver.Chrome()
    wait = WebDriverWait(driver,5)
    
    driver.get(url)
    response = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,'movie-item-info')))
    
    movies = []
    for element in response:
        movie = element.text.split('\n')
        movie[1] = movie[1][3:]
        movie[2] = movie[2][5:]
        movies.append(movie)
    
    driver.quit()
    return movies
```

```python
def main():
    wb = xl.Workbook()
    st = wb.active
    row = 1
    legend = ['名称', '主演', '上映时间']
    
    for col in range(3):
        st.cell(row = row, column = col+1).value = legend[col]
    
    row += 1
    
    for i in range(10):
        url = 'https://maoyan.com/board/4?offset={}'.format(i)
        movies = driver_start(url)
        
        for movie in movies:
            for col in range(3):
                st.cell(row = row, column = col+1).value = movie[col]
            row += 1
            
    wb.save("猫眼电影top100.xlsx")
```

```python
main()
```
